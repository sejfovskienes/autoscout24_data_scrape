```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--headless")

driver = webdriver.Chrome(options=chrome_options)
```

```python
def close_popup():
  try:
    consent_button = driver.find_element(By.CLASS_NAME, "_consent-accept_1lphq_114")
    consent_button.click()
    print("Closed cookie popup")
    time.sleep(2)
  except NoSuchElementException:
      print("No cookie popup found")
```

```python
  url = "https://www.autoscout24.com/"

  driver.get(url)

  time.sleep(5)
```

```python
make_select = Select(driver.find_element(By.ID, "make"))

makes = [option.get_attribute("value") for option in make_select.options if option.get_attribute("value")]

print(makes)
```

```python
all_cars = []
```

```python
print(f"Total cars scraped: {len(all_cars)}")

for make in makes:
  print(f"scraping for brand: {make}")
  make_select = Select(driver.find_element(By.ID, "make"))
  make_select.select_by_value(make)
  time.sleep(2)

  model_select = Select(driver.find_element(By.ID, "model"))
  models = [option.get_attribute("value") for option in model_select.options if option.get_attribute("value")]

  for model in models:
    print(f"scraping for model {model}")
    model_select = Select(driver.find_element(By.ID, "model"))
    model_select.select_by_value(model)
    time.sleep(2)
    close_popup()
    search_button = driver.find_element(By.ID, "search-mask-search-cta")
    search_button.click()
    time.sleep(5)


    while True:
      car_elements = driver.find_elements(By.CLASS_NAME, "ListItem_wrapper__TxHWu")
      all_cars.extend(car_elements)
      print(f"{len(car_elements)} more elements added, total elements scraped: {len(all_cars)}")

      try:
          next_button = driver.find_element(By.XPATH, "//button[@aria-label='Go to next page']")


          if next_button.get_attribute("aria-disabled") == "true":
              print(f" No more pages for {make}-{model}")
              break

          next_button.click()
          time.sleep(5)

      except NoSuchElementException:
          print(f" Finished scraping {make}-{model}")
          break

    driver.get(url)
    time.sleep(5)
```
